## Convert Hinglish into English as an NLP task using the Google Translate library.

```python
from googletrans import Translator

translator = Translator()

text = '''Kal main subah jaldi uth gaya tha kyunki mujhe ek important project complete karna tha. Maine pehle chai banayi aur thodi der balcony mein baith kar weather enjoy kiya. Uske baad maine laptop open kiya aur apna NLP assignment continue kiya jisme mujhe Hinglish text ko English mein convert karna tha using translation libraries. Starting mein kaafi errors aa rahe the because mixed language sentences ko properly detect karna thoda difficult hota hai, lekin baad mein maine preprocessing techniques use ki jaise text normalization aur token cleaning. Afternoon mein mere dost ka call aaya aur humne AI, machine learning, aur future technologies ke baare mein discussion kiya. Evening tak mera kaafi kaam complete ho gaya tha, toh maine thoda music suna aur walk ke liye bahar chala gaya. Raat ko dinner ke baad maine final testing ki aur dekha ki model ka output pehle se kaafi better aa raha tha. Overall aaj ka din productive bhi tha aur interesting bhi because mujhe naye NLP concepts seekhne ko mile.'''
translated = translator.translate(text, src='hi', dest='en')

print(translated.text)
```

```python
from nltk.tokenize import word_tokenize , sent_tokenize
sentences = sent_tokenize(translated.text)

print(sentences)
```
