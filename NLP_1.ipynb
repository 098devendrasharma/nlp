## Step: 2 --> Remove HTML Tags

```python
import re
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'', text)
```

```python
text = "<html><body><p> Movie1 </p><p> Click here to <a href = 'http://google.com'>download</a></p>"
remove_html_tags(text)
```

```python
df['review'] = df['review'].apply(remove_html_tags)
df.head()
```

```python
## If you want to remove url from text
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'', text)
```

```python
from os import remove
text1 = 'For notebook click https://www.deture.com/delta/note123acc'
remove_url(text1)
```

```python
df['review'] = df['review'].apply(remove_url)
df.head()
```

## Step: 3 --> Remove Punctuation

```python
import string
string.punctuation
```

```python
exclude = string.punctuation

def remove_punc(text):
  for char in exclude:
    text = text.replace(char, '')
  return text
```

```python
test = 'string. With. Punctuation?'
remove_punc(test)
```

```python
df['review'] = df['review'].apply(remove_punc)
df
```

## Step: 4 --> Chatword Treatment

```python
chat_words = {"u2" : "you too", "BBL" : "Be Back Later", "GAL" : "Get a life"}
chat_words
```

```python
def chat_conversion(text):
  new_text = []
  for word in text.split():
    if word.upper() in chat_words:
      new_text.append(chat_words[word.upper()])
    else:
      new_text.append(word)
  return "".join(new_text)
```

```python
chat_conversion("BBL")
```

## Step: 6 --> Removing Stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

STOPWORDS = set(stopwords.words('english'))
```

```python
def remove_stopwords(text: str):
  new_text = []
  for word in text.split():
    if word in STOPWORDS:
      new_text.append('')
    else:
      new_text.append(word)

  x = new_text[:]
  new_text.clear()
  return " ".join(x)

remove_stopwords('probably my all-time favourite movie, a story of selflessness, sacrifice and dedicated to a noble cause')
```
